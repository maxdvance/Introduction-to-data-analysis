```python
import pandas as pd
import numpy as np
```

```python
# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
df = pd.read_excel(r'Superstore.xlsx')
```

```python
df
```

```python
import datetime as dt
```

```python
df['year'], df['month'] = df['Order Date'].dt.year, df['Order Date'].dt.month
```

```python
df.groupby('year').agg('sum')['Profit'].plot(x='Year', y='Profit',kind='bar')  
plt.title('Profit vs Year')  
plt.xlabel('Year')  
plt.ylabel('Profit')  
plt.show()
```

```python
df.groupby(['year','month']).agg('sum')['Profit'].unstack(level=0).plot()  
plt.title('Profit vs Year')  
plt.xlabel('Year')  
plt.ylabel('Profit')  
plt.show()
```
